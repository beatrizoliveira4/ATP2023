```python
### TPC4


from random import randint

menu = print('''
    Menu:
    * (1) Criar Lista 
    * (2) Ler Lista
    * (3) Soma
    * (4) Média
    * (5) Maior
    * (6) Menor
    * (7) estaOrdenada por ordem crescente
    * (8) estaOrdenada por ordem decrescente
    * (9) Procura um elemento
    * (0) Sair
    ''')

n = int(input("Escolha a opção do menu que pretende efetuar: "))

lista = []

while n != 0:

    if n == 1:

        lista.clear()
        N = int(input("Quantos números quer que a sua lista contenha? "))
        while N > 0:
            lista.append(randint(1,100))
            N = N - 1
        print ("A sua lista é " + str(lista))
        n = int(input("Insira a opção que deseja: "))

    if n == 2:

        lista.clear() 
        N = int(input("Quantos números quer que a sua lista contenha? "))
        while N > 0:
            num = int(input("Insira um número para a lista: "))
            lista.append(num)
            N = N - 1
        print ("A sua lista é " + str(lista))
        n = int(input("Insira a opção que deseja: "))

    if n == 3:
        soma = 0
        for x in lista:
            soma = soma + x
        print ("A soma dos elementos da lista é " + str(soma))
        n = int(input("Insira a opção que deseja: "))

    if n == 4:
        soma = 0
        for x in lista:
            soma = soma + x
            média = soma/len(lista)
        print ("A média dos elementos da lista é " + str(média))
        n = int(input("Insira a opção que deseja: "))
    
    if n == 5:
        maior = lista[0] 
        for x in lista[1:]:
            if x > maior:
                maior = x 
        print ("O maior elemento da lista é " + str(maior))
        n = int(input("Insira a opção que deseja: "))

    if n == 6:
        menor = lista[0] 
        for x in lista[1:]:
            if x < menor:
                menor = x 
        print ("O menor elemento da lista é " + str(menor))
        n = int(input("Insira a opção que deseja: "))
        
    if n == 7:
        tam = len(lista)
        ordem = "Sim!"
        while tam > 0:
            for x in range(0, tam - 1):
                if lista[x] > lista[x + 1]:
                    ordem = "Não!"
            tam = tam - 1
        print("A lista está ordenada por ordem crescente? - " + str(ordem))
        n = int(input("Insira a opção que deseja: "))
    
    if n == 8:
        tam = len(lista)
        ordem = "Sim!"
        while tam > 0:
            for x in range(0, tam - 1):
                if lista[x] < lista[x + 1]:
                    ordem = "Não!"
            tam = tam - 1
        print("A lista está ordenada por ordem decrescente? - " + str(ordem))
        n = int(input("Insira a opção que deseja: "))

    if n == 9:
        elem = int(input("Qual o elemento que quer procurar na lista? "))
        if elem in lista:
            i = lista.index(elem)
        elif elem not in lista:
            i = -1
        print("O índice do elemento na lista é " + str(i+1))
        n = int(input("Insira a opção que deseja: "))
        
if n == 0: 
    print("Saiu do programa! Esperamos que tenha sido útil")
```
